```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType,StructField, StringType, IntegerType, DoubleType, TimestampType, DateType


!wget https://github.com/databricks/spark-xml/raw/master/src/test/resources/books.xml  -P /content/sample_data/

spark = SparkSession.builder.master("local[*]").appName("xmlmami") \
  .config('spark.jars.packages', 'com.databricks:spark-xml_2.12:0.6.0') \
  .getOrCreate()

schema = StructType([ \
    StructField("_id", StringType(), False), \
    StructField("author", StringType(), False), \
    StructField("description", StringType(), True), \
    StructField("genre", StringType(), False), \
    StructField("price", DoubleType(), False), \
    StructField("publish_date", DateType(), False), \
    StructField("title", StringType(), False)
])  

df = spark.read \
          .format('com.databricks.spark.xml') \
          .options(rowTag='book') \
          .option("inferSchema", "true") \
          .option("dateFormat", "yyyy/MM/dd") \
          .schema(schema) \
          .load('sample_data/books.xml')

df = df.withColumnRenamed("_id","Id")

df.show(n=50,truncate=False)
```

```python


df1 = spark.read \
          .format('com.databricks.spark.xml') \
          .options(rootTag='postlinks') \
          .options(rowTag='row') \
          .load('sample_data/Comments.xml')

df1.printSchema()

df1.show(n=120, truncate=False)

df1.write.partitionBy('_PostId').mode('overwrite').parquet("sample_data/comment.parquet") 
```

```python
postsDF = spark.read \
          .format('com.databricks.spark.xml') \
          .options(rootTag='posts') \
          .options(rowTag='row') \
          .load('sample_data/Posts.xml')

postsDF.printSchema()

postsDF.show(n=120, truncate=False)

postsDF.write.partitionBy('_PostTypeId').mode('overwrite').parquet("sample_data/post.parquet") 
```
